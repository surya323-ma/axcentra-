```python
# -----------------------------------------------------
# COVID-19 EDA (Using total_cases, total_deaths, total_vaccinations)
# -----------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Load dataset
df = pd.read_csv("covid_19_clean_complete.csv")

# Convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# -----------------------------------------------------
# 2️⃣ Trend Line: total_cases vs date (Top 5 countries)
# -----------------------------------------------------

# Top 5 countries with highest total cases
top5 = df.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).head(5).index

df_top5 = df[df['Country/Region'].isin(top5)]

plt.figure(figsize=(12,6))

for country in top5:
    country_data = df_top5[df_top5['Country/Region'] == country]
    plt.plot(country_data['Date'], country_data['Confirmed'], label=country)

plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.title("COVID-19 Trend Line – Top 5 Countries")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

3️⃣ Compare Top 5 Countries (Bar Chart)

```python

top5_cases = df.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
plt.bar(top5_cases.index, top5_cases.values)
plt.xlabel("Country")
plt.ylabel("Total Cases")
plt.title("Top 5 Countries with Highest COVID-19 Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4️⃣ Heatmap (Correlation)

```python
corr = df[['Confirmed','Deaths','Recovered']].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

5️⃣ Scatter Plot: total_cases vs total_deaths

```python

plt.figure(figsize=(7,5))
plt.scatter(df['Confirmed'], df['Deaths'], alpha=0.5)
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.title("Scatter Plot: Cases vs Deaths")
plt.grid(True)
plt.tight_layout()
plt.show()
```
