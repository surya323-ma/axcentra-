```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

1.Create a pandas **DataFrame** manually.

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))
print(df.shape)
df.head()
```

```python
clean_df = df.dropna()
```

2. Handle missing values using **fillna()**

```python
df_filled = df.fillna({
    "math score": df["math score"].mean(),
    "reading score": df["reading score"].median()
})

print("DataFrame after fillna():")
print(df_filled.head(), "\n")
```

3. Sort DataFrame by a column

```python
df_sorted = df_filled.sort_values(by="math score", ascending=False)
print("DataFrame sorted by Math Score (descending):")
print(df_sorted, "\n")
```

4. Find mean, median, mode of numeric columns

```python
mean_values = df_filled[["math score", "reading score","writing score"]].mean()
median_values = df_filled[["math score", "reading score","writing score"]].median()
mode_values = df_filled[["math score", "reading score","writing score"]].mode()

print("Mean Values:")
print(mean_values, "\n")

print("Median Values:")
print(median_values, "\n")

print("Mode Values:")
print(mode_values)
```

5.Calculate average, max, min scores

```python
# Calculate score statistics
results = pd.DataFrame({
    "Average": [
        clean_df["math score"].mean(),
        clean_df["reading score"].mean(),
        clean_df["writing score"].mean()
    ],
    "Max": [
        clean_df["math score"].max(),
        clean_df["reading score"].max(),
        clean_df["writing score"].max()
    ],
    "Min": [
        clean_df["math score"].min(),
        clean_df["reading score"].min(),
        clean_df["writing score"].min()
    ]
}, index=["Math Score", "Reading Score", "Writing Score"])

# Display the results table
print(results)
```
